```python
from transformers import PegasusForConditionalGeneration, PegasusTokenizer
```

```python
model_name = "ailm/pegsus-text-summarization"
model = PegasusForConditionalGeneration.from_pretrained(model_name)
tokenizer = PegasusTokenizer.from_pretrained(model_name)
```

```python
text = """Harry Potter is a series of seven fantasy novels written by British author J. K. Rowling.
The novels chronicle the lives of a young wizard, Harry Potter, and his friends, Hermione Granger and
Ron Weasley, all of whom are students at Hogwarts School of Witchcraft and Wizardry. The main story arc
concerns Harry's conflict with Lord Voldemort, a dark wizard who intends to become immortal, overthrow the
wizard governing body known as the Ministry of Magic, and subjugate all wizards and Muggles (non-magical people)."""
```

```python
tokens = tokenizer(text, truncation=True, padding="longest", return_tensors="pt")
```

```python
summary = model.generate(**tokens)
print(tokenizer.decode(summary[0], skip_special_tokens=True))
```
